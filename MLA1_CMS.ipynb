# **Rename Columns**
rename columns to date_time, temp_c, dew_point, relative_humidity, wind_speed, visibility, pressure, and weather

```python
df.rename(columns = {'Date/Time': 'date_time', 'Temp (C)': 'temp_c','Dew Point Temp (C)':'dew_point','Rel Hum (%)':'relative_humidity', 'Wind Spd (km/h)': 'wind_speed','Visibility (km)':'visibility','Stn Press (kPa)':'pressure','Weather':'weather'},inplace=True)

df.head()
```

# **Standard Statistics of each Column**
This code will provide standard stats(mean,count, min/max,etc.) for the numerical data

```python
df.describe()
```

# **Add new column for Fahrenheit**
This code will add a new column for F, and use the C column to convert to F

```python
df['temp_f'] = df['temp_c'] * 9/5 + 32

df.head()
```

# **Rearrange the Columns of the Dataset**
This code will make temp_f appear right after temp_c

```python
newOrder = ['date_time','temp_c','temp_f','dew_point','relative_humidity','wind_speed','visibility','pressure','weather']

df = df[newOrder]
df.head()
```

# **Display Data Types of all Columns**

```python
df.dtypes
```

# **Convert the date_time column**
This code will convert the date_time column to reflect date/time instead of an object

```python
df['date_time'] = pd.to_datetime(df['date_time'])
df.dtypes
```

# **Get the Avg. Temp for each Month**
This code will create an object called monthly_avg

```python
df['monthly_avg'] = df['date_time'].dt.month

monthly_avg = df.groupby('monthly_avg')['temp_f'].mean()

print(monthly_avg)
```

# **Scatterplot for Temp(F) vs Humidity**

```python
df.plot(kind='scatter',x='temp_f',y='relative_humidity',figsize=(20,10), xlabel='Tempature (F)',ylabel='Relative Humidity',title='Tempature (F) vs Relative Humidity')
```

# **Line Chart for Avg. Monthly Temp**

```python
monthly_avg.plot(kind='line', x='temp_f',y='monthly_avg',figsize=(20,10), xlabel="Month", ylabel="Tempature (F)",title="Average Tempature per Month")
```
